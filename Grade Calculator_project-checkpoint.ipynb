```python
#  import the required libraries
import pandas
import matplotlib
```

```python
def get_grade(scores):          # Defining the function
    '''
        A function that takes student scores as input, and outputs grades.
        Use the following grading criteria for input scores
        100 - 70 = A
        69 - 65 = B
        64 - 55 = C
        54 - 45 = D
        44 - 0 = F
        Requirements
        1. The function should be able to accept multiple scores as input and print out the corresponding number of grades
        2. The use of control flow must be demonstrated
        3. Adequate documentation
    '''
    grades= []    # initializing empty list to store grades
   
    # Validating input to ensure it is within recommended range through loop
    for score in scores:
        if score > 100 or score < 0:                
            grades.append("invalid")
        elif score >= 70:
            grades.append("A")
        elif score >= 65:
            grades.append("B")
        elif score >= 55:
            grades.append("C")
        elif score >= 45:
            grades.append("D")
        else:
            grades.append("F")
    return grades

# Accepts scores as input
scores_input= input("Enter your score: ") 

# Convert the input string to a list
scores_list= [int(score) for score in scores_input.split()]

grades_output= get_grade(scores_list)      # Calling the function to calculate the grades

# Displaying the results 
print("\nScores and Grades")
for i in range(len(scores_list)):
    print(f"Score: {scores_list[i]} = Grade: {grades_output[i]}")
```
